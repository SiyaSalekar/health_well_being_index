```python
# Create a new binary column based on the "Status" column
data['Country_Status'] = data['Status'].apply(lambda x: 1 if x == 'Developed' else 0)

# Drop the original "Status" column 
data.drop('Status', axis=1, inplace=True)
print(data.columns)
```

```python
#check missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
#focusing on only european countries
# List of European countries
european_countries = [
    "Albania", "Armenia", "Austria", "Azerbaijan", "Belarus", "Belgium", "Bosnia and Herzegovina", "Bulgaria",
    "Croatia", "Cyprus", "Czechia", "Denmark", "Estonia", "Finland", "France", "Georgia", "Germany", "Greece",
    "Hungary", "Iceland", "Ireland", "Italy", "Kazakhstan", "Kosovo", "Latvia", "Lithuania", "Luxembourg", "Malta",
    "Moldova", "Monaco", "Montenegro", "Netherlands", "North Macedonia", "Norway", "Poland", "Portugal", "Romania",
    "Russia", "San Marino", "Serbia", "Slovakia", "Slovenia", "Spain", "Sweden", "Switzerland", "Turkey", "Ukraine",
    "United Kingdom", "Vatican City"
]

# Filter out European countries
data = data[data['Country'].isin(european_countries)]
```

```python
# cleaning data

# Check if the country has data for each year from 2000 to 2015

countries_to_delete = []  
for country in data['Country'].unique():
    years = data[data['Country'] == country]['Year'].unique()
    if set(range(2000, 2016)).difference(years):
        print(f"{country} is missing data for the following years:", end=" ")
        missing_years = set(range(2000, 2016)).difference(years)
        print(missing_years)
        countries_to_delete.append(country)
```

```python
#delete above country rows since they have data only for the year 2013
data = data[~data['Country'].isin(countries_to_delete)]
```

```python
#POPULATION MISSING VALUES

#read population data and fill missing values for current data
file_path = 'C:/Users/SIYA/Downloads/EDU_DEM_03052024202813726.csv'
population_data = pd.read_csv(file_path)
```

```python
# Function to fill missing population values
def fill_missing_population(row):
    year = row['Year']
    country = row['Country']
    population_value = row['Population']
    if pd.isnull(population_value):  # Check if the population value is missing
        matching_population = population_data[(population_data['Country'] == country) & (population_data['YEAR'] == year)]['Value'].values
        if len(matching_population) > 0:
            return matching_population[0]
    return population_value  # Return the original value if it's not missing

# Apply the function to fill missing values in the Population column
data['Population'] = data.apply(fill_missing_population, axis=1)

print(data)
```

```python
# for countries not in population data fill using dict obtained from - https://www.macrotrends.net/global-metrics/countries/SGP/<country-name>/population 
# replace <country-name> with country name

# Define a dictionary to map countries to their population data
population_dict = {
    'Czechia': {
        2009: 10421510,
        2008: 10378583,
        2007: 10337829,
        2006: 10304815,
        2004: 10260999,
        2003: 10248038,
        2002: 10238170,
        2001: 10230276
    },
    'Slovakia': {
        2015: 5424444,
        2014: 5419569,
        2013: 5414739,
        2012: 5409283,
        2011: 5402680,
        2010: 5396424,
        2009: 5388926,
        2008: 5381636,
        2007: 5378382,
        2006: 5377235,
        2005: 5376242,
        2004: 5374931,
        2003: 5374376,
        2002: 5375206,
        2001: 5376456,
        2000: 5376690
    }
}

# Fill missing population values based on the dictionary
for index, row in data.iterrows():
    if pd.isnull(row['Population']):
        country = row['Country']
        year = row['Year']
        if country in population_dict and year in population_dict[country]:
            data.at[index, 'Population'] = population_dict[country][year]
```

```python
#check missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
#Hepatitis B MISSING VALUES

# countries with all missing Hepatitis B values
countries_with_missing_hep_b = set()

def fill_missing_hepatitis_b(row):
    global countries_with_missing_hep_b
    country = row['Country']
    hep_b_values = data[data['Country'] == country]['Hepatitis B']
    if hep_b_values.isnull().all() and country not in countries_with_missing_hep_b:
        print(f"All Hepatitis B values missing for {country}")
        countries_with_missing_hep_b.add(country)
    elif not hep_b_values.isnull().all():
        country_mean = hep_b_values.mean()
        row['Hepatitis B'] = country_mean
    return row

#fill missing values in the Hepatitis B column
data = data.apply(fill_missing_hepatitis_b, axis=1)

```

```python
#for countries with missing values for all years for hepatits B fill with avg
#found from - https://med.stanford.edu/liver/research/global-vaccination-coverage.html 

# Manually set Hepatitis B values for specific countries
hepatitis_b_values = {
    'Finland': 2,
    'Denmark': 2,
    'Hungary': 2,
    'Iceland': 2,
    'Norway': 96,
    'Slovenia': 2,
    'Switzerland': 72
}
data.loc[data['Country'].isin(hepatitis_b_values.keys()), 'Hepatitis B'] = data['Country'].map(hepatitis_b_values)
```

```python
def fill_missing_alcohol(row):
    country = row['Country']
    alcohol_values = data[(data['Country'] == country) & data['Alcohol'].notna()]['Alcohol']
    if alcohol_values.any():  # Check if any non-missing values exist for the country
        country_mean = alcohol_values.mean()
        row['Alcohol'] = row['Alcohol'] if not pd.isna(row['Alcohol']) else country_mean
    return row

# Apply the function to fill missing values in the Alcohol column
data = data.apply(fill_missing_alcohol, axis=1)

# Check missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)

# Check if conversion was successful
print(data['Alcohol'].dtype)
```

```python
# replace missing values for Total expenditure, Polio, GDP, Diphtheria, Income composition of resources, Schooling with mean of each column since very small amount of data missing (<35 rows)      

columns_to_fill = ['Total expenditure', 'Polio', 'GDP', 'Diphtheria ', 'Income composition of resources', 'Schooling']

# Iterate through each column and fill missing values with column mean
for col in columns_to_fill:
    data[col].fillna(data[col].mean(), inplace=True)

```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Define columns to scale
columns_to_scale = [col for col in data.columns if col not in ['Country', 'Year', 'Country_Status',' HIV/AIDS']]

# Fit and transform each selected column
for col in columns_to_scale:
    scaler.fit(data[[col]])
    data[col] = scaler.transform(data[col].values.reshape(-1, 1))
```

```python
#Life Expectancy Outliers 
# Using Inter-Quartile Method
Q1 = data['Life_expectancy'].quantile(0.25)
Q3 = data['Life_expectancy'].quantile(0.75)
# Calculate IQR
IQR = Q3 - Q1
# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Identify outliers
outliers = data[(data['Life_expectancy'] < lower_bound) | (data['Life_expectancy'] > upper_bound)]
#printing outliers to see if not all data(for all years) for a country is deleted
print(outliers)
# Remove outliers
data = data[(data['Life_expectancy'] >= lower_bound) & (data['Life_expectancy'] <= upper_bound)]
```

Principal Component Analysis

```python
#keep unique countries and replace column values by average over the years
average_data = data.groupby('Country').mean()
average_data = average_data.drop(columns=['Year'])
df = pd.DataFrame(average_data).reset_index()
print(df)

#We observe total 40 countries stored in the format requireed to perform PCA
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

# Extracting numerical columns
numerical_columns = df.columns[1:]

# Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[numerical_columns])

# Perform PCA
pca = PCA()
pca.fit(scaled_data)

# Compute Eigenvectors
eigenvectors = pca.components_

# Select the number of components based on explained variance
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance_ratio.cumsum()

# Choose the number of components that explain a desired amount of variance
desired_variance_explained = 0.95  # You can adjust this threshold
n_components = (cumulative_explained_variance < desired_variance_explained).sum() + 1

# Perform PCA with the selected number of components
pca = PCA(n_components=n_components)
pca_data = pca.fit_transform(scaled_data)

# Create DataFrame for the PCA results
pca_columns = [f"PC{i+1}" for i in range(n_components)]
pca_df = pd.DataFrame(data=pca_data, columns=pca_columns)

# Concatenate with the country column
pca_df = pd.concat([df[['Country']], pca_df], axis=1)

# Interpret Eigenvectors
eigenvectors_df = pd.DataFrame(eigenvectors, columns=numerical_columns)

# Print Eigenvectors
print("Eigenvectors:")
print(eigenvectors_df)

# Features to include in the biplot
features_to_plot = ['Life_expectancy', 'Adult_Mortality', 'Alcohol']

# Plot biplot
plt.figure(figsize=(10, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], alpha=0.6)

# Plot eigenvectors for selected features as arrows
for feature in features_to_plot:
    feature_index = numerical_columns.get_loc(feature)
    comp = eigenvectors[:, feature_index]
    var = pca.explained_variance_[0]
    plt.arrow(0, 0, comp[0]*np.sqrt(var), comp[1]*np.sqrt(var), 
              head_width=0.1, head_length=0.1, color='r', alpha=0.75)
    plt.text(comp[0]*np.sqrt(var)*1.2, comp[1]*np.sqrt(var)*1.2, feature, color='g')

# Features to include in the biplot
features_to_plot = ['infant_deaths', 'Hepatitis_B', 'Polio']

# Plot biplot
plt.figure(figsize=(10, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], alpha=0.6)

# Plot eigenvectors for selected features as arrows
for feature in features_to_plot:
    feature_index = numerical_columns.get_loc(feature)
    comp = eigenvectors[:, feature_index]
    var = pca.explained_variance_[0]
    plt.arrow(0, 0, comp[0]*np.sqrt(var), comp[1]*np.sqrt(var), 
              head_width=0.1, head_length=0.1, color='r', alpha=0.75)
    plt.text(comp[0]*np.sqrt(var)*1.2, comp[1]*np.sqrt(var)*1.2, feature, color='g')

# Set labels
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Biplot')

# Show plot
plt.grid()
plt.show()
```

From the above Bi-plots we understand that:
<ul>
<li>
The arrows for life expectancy and adult mortality are in opposite directions, indicating a negative correlation between them. This suggests that countries with higher life expectancy tend to have lower adult mortality rates.</li>
<li>
The vector representing alcohol is shorter compared to the vectors representing life expectancy and adult mortality, it indicates that alcohol consumption has lower variance compared to them, with life expectancy having highest variance comparitively.
</li>
<li> From the second bi-plot we observe Polio and Hepatitis B have very lower variance.
</li>
</ul>

CLUSTERING
<br><br>
I chose Hierarchial Clustering because
<ul>
<li>No need to specify the number of clusters: Hierarchical clustering builds a tree-like structure (dendrogram) that allows you to visualize the clustering process at different levels of granularity which is useful in this case as we are unsure about the appropriate number of clusters in the data.</li>
<li>Robustness to outliers: My data includes some outliers and Hierarchical clustering can handle outliers gracefully, since it's based on the distance between data points, outliers typically don't affect the overall clustering structure as much as in other methods like k-means. </li>
<li>Interpretability: The dendrogram produced by hierarchical clustering provides a clear visual representation of how the data points are grouped together, which helps to identify structure of data.</li>
<li>Flexibility in linkage methods: Hierarchical clustering allows to choose different linkage methods (e.g., single, complete, average, etc.) depending on desired clustering outcome.</li>
<li>Suitable for small datasets: Since hierarchical clustering compares all pairs of data points, it can be computationally expensive for large datasets. However, for small to moderate-sized datasets, the computational overhead is typically manageable.</li></ul>


```python
import scipy.cluster.hierarchy as sch

X_subset = df['Life_expectancy'].values.reshape(-1, 1)
labels = df['Country']

plt.figure(figsize=(6, 14))
dendrogram = sch.dendrogram(sch.linkage(X_subset, method='complete',metric='euclidean'), labels=labels.values)
plt.title('Life expectancy Hierarchical Clustering Dendrogram')
plt.xlabel('Countries')
plt.ylabel('Distance')
plt.show()
```

I used complete linkage because it resulted in relatively compact and well-separated clusters. Also the maximum distance for single and average is 0.3 while for complete max distance of clustering is 0.6 hence prioritizing variance and maintaining distinctiveness.

WEIGHING AND AGGREGATION - INDEX CREATION
<br>
Based on loading scores obtained from PCA
We assign: <br><i>Income_composition_of_resources, thinness_5_9_years, Life_expectancy, GDP</i> <br>comparitively more weightage compared to other variables. 

```python
import pandas as pd

# Loading scores for each variable
loading_scores = {
    'Income_composition_of_resources': 0.378418,
    'thinness_5_9_years': 0.367352,
    'Life_expectancy': 0.364571,
    'GDP': 0.322913,
    'Adult_Mortality': 0.293738,
    'Country_Status': 0.285518,
    'infant_deaths': 0.260875,
    'Measles': 0.252820,
    'Alcohol': 0.235054,
    'Polio': 0.224361,
    'Total_expenditure': 0.206979,
    'HIV/AIDS': 0.159822,
    'Hepatitis_B': 0.094404
}

# Normalize weights
total_weight = sum(loading_scores.values())
normalized_weights = {key: value / total_weight for key, value in loading_scores.items()}

# Printing normalized weights
print("Normalized Weights:")
for indicator, weight in normalized_weights.items():
    print(f"{indicator}: {weight}")

# Weighting each indicator for each country
weighted_indicators = pd.DataFrame()
for col in normalized_weights:
    weighted_indicators[col] = df[col] * normalized_weights[col]

# Creating sub-indicators for each country
sub_indicators = pd.DataFrame()
sub_indicators['Country'] = df['Country']  # Adding country column at first
sub_indicators['Economic_Status'] = (weighted_indicators['Income_composition_of_resources'] + 
                                            weighted_indicators['GDP'] + 
                                            weighted_indicators['Total_expenditure'])
sub_indicators['Mortality'] = (weighted_indicators['Adult_Mortality'] + 
                                          weighted_indicators['infant_deaths'])
sub_indicators['Disease_Burden'] = (weighted_indicators['Measles'] + 
                               weighted_indicators['Polio'] + 
                               weighted_indicators['HIV/AIDS'] + 
                               weighted_indicators['Hepatitis_B'])
sub_indicators['Health_and_Lifestyle'] = (weighted_indicators['Life_expectancy'] + 
                                          weighted_indicators['Alcohol'] + 
                                          weighted_indicators['thinness_5_9_years'])
sub_indicators['Development_Status'] = weighted_indicators['Country_Status']

# Calculating the average of each sub-indicator for each country
sub_indicators['Health_quality_index'] = sub_indicators.drop(columns=['Country']).sum(axis=1)

print(sub_indicators)
```

Observations:

<li>We observe that countries with higher scores in economic status, such as Luxembourg, Switzerland, and Ireland, tend to have higher health quality indices.</li>
<li>Countries with lower mortality rates, such as Sweden and Denmark, tend to have higher health quality indices.</li>
<li>Countries with lower disease burdens, such as Iceland and Finland, tend to have higher health quality indices.</li>
